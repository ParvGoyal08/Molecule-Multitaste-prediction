```python
# Data Analysis for Report
import pandas as pd
import numpy as np

# Load the cleaned dataset
df = pd.read_csv("chemtastesdb_multilabel_clean.csv")
print(f"Total samples: {len(df)}")

# Class distribution
print("\n=== CLASS DISTRIBUTION ===")
labels = ['Sweet', 'Bitter', 'Umami', 'Sour', 'Salty']
for label in labels:
    count = df[label].sum()
    pct = count / len(df) * 100
    print(f"{label}: {count} ({pct:.1f}%)")

# Neutral/Tasteless (num_tastes == 0)
neutral = (df['num_tastes'] == 0).sum()
print(f"\nNeutral/Tasteless (no taste label): {neutral} ({neutral/len(df)*100:.1f}%)")

# Multi-label distribution
print("\n=== MULTI-LABEL DISTRIBUTION ===")
for i in range(0, 6):
    count = (df['num_tastes'] == i).sum()
    pct = count / len(df) * 100
    print(f"{i} tastes: {count} ({pct:.1f}%)")

# Multi-label (more than 1 taste)
multi_label = (df['num_tastes'] > 1).sum()
print(f"\nMulti-label samples (>1 taste): {multi_label} ({multi_label/len(df)*100:.1f}%)")

# Single-label
single_label = (df['num_tastes'] == 1).sum()
print(f"Single-label samples (exactly 1 taste): {single_label} ({single_label/len(df)*100:.1f}%)")
```

```python
# Original data before cleaning
import pandas as pd

# Load original Excel file
df_original = pd.read_excel("ChemTastesDB_database.xlsx")
print(f"Original dataset: {len(df_original)} rows")
print(f"Rows with missing values: {df_original.isna().any(axis=1).sum()}")
print(f"After dropping missing: {len(df_original.dropna())} rows")

# Check duplicate SMILES in original
smiles_counts = df_original['canonical SMILES'].value_counts()
duplicated = (smiles_counts > 1).sum()
print(f"\nDuplicate SMILES entries: {duplicated}")
print(f"Total rows involved in duplicates: {smiles_counts[smiles_counts > 1].sum()}")

# Final cleaned dataset
df_clean = pd.read_csv("chemtastesdb_multilabel_clean.csv")
print(f"\nFinal unique molecules (after duplicate merge): {len(df_clean)}")
```
